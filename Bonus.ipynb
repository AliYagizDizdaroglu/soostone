```python
# Step 2: Start H2O
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
# Step 3: Import Data
data_path = "nyc-rolling-sales.csv"  # Update this path to your dataset's location
data = h2o.import_file(path=data_path)

# Assuming 'SALE PRICE' is your target column
data["SALE PRICE"] = data["SALE PRICE"].asnumeric()

# Identify feature and target columns (assuming you have done so)
feature_columns = data.columns[:-1]  # Update this list with your actual feature column names
target_column = "SALE PRICE"  # Update this with your actual target column name

# Step 3: Split Data
train, valid, test = data.split_frame(ratios=[.7, .15], seed=1234)

# Step 4: Run AutoML with RMSE as the sort metric
aml = H2OAutoML(max_models=10,
                max_runtime_secs=3600,
                seed=1,
                sort_metric='RMSE')  # Focusing on RMSE
aml.train(x=feature_columns, y=target_column, training_frame=train, leaderboard_frame=valid)

# Step 5: View the Leaderboard
print(aml.leaderboard)

# Step 6: Evaluate the Best Model on the Test Set
leader_model = aml.leader
test_performance = leader_model.model_performance(test)
print("RMSE:", test_performance.rmse())
print("R2:", test_performance.r2())

# Step 7: Make Predictions
predictions = leader_model.predict(test)
print(predictions.head())
```

```python
# Access feature importance for the leader model
feature_importance = leader_model.varimp()

# Display feature importance
print("Feature Importance:")
print(feature_importance)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
file_path = 'data_cleaned.csv'
data_raw = pd.read_csv(file_path)
data = data_raw.copy()
```

```python
sns.countplot(x='BOROUGH', data=data, order=data['BOROUGH'].value_counts().index)
plt.title('Sales Distribution by Borough')
plt.ylabel('Number of Sales')
plt.xlabel('Borough')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x='BOROUGH', y='SALE PRICE', data=data, estimator=np.mean, ci=None, order=data.groupby('BOROUGH')['SALE PRICE'].mean().sort_values(ascending=False).index)
plt.title('Average Sale Price by Borough')
plt.ylabel('Average Sale Price')
plt.xlabel('Borough')
plt.xticks(rotation=45)
plt.show()
```

```python
data_grouped = data.groupby('BOROUGH').sum()[['RESIDENTIAL UNITS', 'COMMERCIAL UNITS']]
data_grouped.plot(kind='bar', stacked=True)
plt.title('Residential vs. Commercial Units Sold by Borough')
plt.ylabel('Number of Units Sold')
plt.xlabel('Borough')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.histplot(data['AGE OF PROPERTY'], bins=30, kde=True)
plt.title('Property Age Distribution')
plt.xlabel('Age of Property (Years)')
plt.ylabel('Frequency')
plt.show()
```

```python
sales_monthly = data.groupby('SALE YEAR_MONTH').size()
sales_monthly.plot(kind='line', marker='o')
plt.title('Monthly Sales Volume Over Time')
plt.xlabel('Year and Month')
plt.ylabel('Number of Sales')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
average_price_monthly = data.groupby('SALE YEAR_MONTH')['SALE PRICE'].mean()
average_price_monthly.plot(kind='line', marker='o', color='green')
plt.title('Average Monthly Sale Price Over Time')
plt.xlabel('Year and Month')
plt.ylabel('Average Sale Price')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='BUILDING CLASS AT TIME OF SALE', data=data, order=data['BUILDING CLASS AT TIME OF SALE'].value_counts().index[:10])
plt.title('Top 10 Building Class Distribution at Time of Sale')
plt.ylabel('Number of Sales')
plt.xlabel('Building Class at Time of Sale')
plt.xticks(rotation=45)
plt.show()
```
